```python
numeric_variables = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_variables.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```

```python
numeric_variable1 = 'Ex-Showroom_Price'
numeric_variable2 = 'Cylinders'
sns.jointplot(x=numeric_variable1, y=numeric_variable2, data=df, kind='scatter')
plt.suptitle(f'Joint Plot of {numeric_variable1} and {numeric_variable2}', y=1.02)
plt.show()

```

```python

categorical_variable1 = 'Drivetrain'
categorical_variable2 = 'Fuel_Type'

stacked_df = df.groupby([categorical_variable1, categorical_variable2]).size().unstack()
stacked_df.plot(kind='bar', stacked=True, figsize=(12, 4))
plt.title('Stacked Bar Chart of Categorical Variables')
plt.xlabel(categorical_variable1)
plt.ylabel('Count')
plt.show()
```

```python
categorical_variable = 'Drivetrain'
numeric_variable = 'Ex-Showroom_Price'

# Create a violin plot
plt.figure(figsize=(12, 8))
sns.violinplot(x=categorical_variable, y=numeric_variable, data=df)
plt.title(f'Violin Plot of {numeric_variable} across {categorical_variable}')
plt.xlabel(categorical_variable)
plt.ylabel(numeric_variable)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.catplot(x=categorical_variable, y=numeric_variable, data=df, kind='bar')
plt.title(f'Categorical Plot of {numeric_variable} by {categorical_variable}')
plt.xlabel(categorical_variable)
plt.ylabel(numeric_variable)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.show()
```

```python
numeric_variable = 'Ex-Showroom_Price'
categorical_variable = 'Drivetrain'
hue_variable = 'Fuel_Type'

# Create a box plot with hue
plt.figure(figsize=(12, 8))
sns.boxplot(x=categorical_variable, y=numeric_variable, hue=hue_variable, data=df)
plt.title(f'Box Plot of {numeric_variable} by {categorical_variable} with Hue')
plt.xlabel(categorical_variable)
plt.ylabel(numeric_variable)
plt.legend(title=hue_variable, bbox_to_anchor=(1, 1))
plt.show()
```

```python

manufacturer_column_name = 'Make'

manufacturer_counts = df[manufacturer_column_name].value_counts()


plt.figure(figsize=(12, 10))
patches, texts, autotexts = plt.pie(manufacturer_counts, labels=manufacturer_counts.index, autopct='%1.1f%%', startangle=90)


legend_labels = [f'{manufacturer}: {count}' for manufacturer, count in zip(manufacturer_counts.index, manufacturer_counts)]
plt.legend(patches, legend_labels, title='MANUFACTURES', loc='center left', bbox_to_anchor=(1.2, 0.5))

for autotext in autotexts:
    autotext.set_color('white')

plt.title('Distribution of Car Manufacturers')
plt.show()

```

```python

categorical_variable = 'Drivetrain'


plt.figure(figsize=(12, 8))
sns.countplot(x=categorical_variable, data=df, palette='muted')

plt.title(f'Count Plot of {categorical_variable}')
plt.xlabel(categorical_variable)
plt.ylabel('Count')

plt.xticks(rotation=45)

plt.show()
```
